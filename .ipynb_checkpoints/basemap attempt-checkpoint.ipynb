```python
import matplotlib.pyplot as plt
import matplotlib.cm
import mpl_toolkits
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import Normalize
import pandas as pd
import numpy as np

```

```python
data = "SuicideRiskRateTestSample.csv"
df = pd.read_csv(data)
yes = ["Arizona", "Colorado", "Utah", "Wyoming"] 
compound_total = df.groupby("State")["Compound"].sum()


ss = {"Suicide": 0.0, "Non":0.0}
a = 0
b = 0
for each in compound_total.index:
    if each in yes:
        ss["Suicide"] = ss["Suicide"] + compound_total[each]
    else:
        ss["Non"] = ss["Non"] + compound_total[each]
```

```python
df2 = pd.read_csv("Zipcode_States_With_Geo_Cordinates.csv")
lat = df2.groupby("State")["Latitude"].mean()
lon = df2.groupby("State")["Longitude"].mean()
dfff = {"lat":[], "lon":[]}
for each in lat.index:
    dfff["lat"].append(lat[each])
    dfff["lon"].append(lon[each])
hah = pd.DataFrame(dfff)
n = hah.set_index([lat.index])
n
```

```python

m = Basemap(resolution='c', # c, l, i, h, f or None
            projection='merc',
            #lat_0=54.5, lon_0=-4.36,
            llcrnrlon=-124.62, llcrnrlat= 25.88, urcrnrlon=-67.32, urcrnrlat=48.04)

m.drawmapboundary(fill_color='#46bcec')
m.fillcontinents(color='#f2f2f2',lake_color='#46bcec')
m.drawcoastlines()
for n in range(10**6, 3*10**6, 10**5):
    m.plot(n*1.5,n,'o',markersize=6,color = '#444444', alpha=.8)  
plt.show()
```
