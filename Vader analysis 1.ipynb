```python
# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
import pandas as pd
import numpy as np

```

```python
# Sample Strings
file = "Tweet_Dump.csv"
df = pd.read_csv(file)
# Target String Setting
df.head()
```

```python

# Run analysis
for index, row in df.iterrows():
    compound = analyzer.polarity_scores(row["text"])["compound"]
    pos = analyzer.polarity_scores(row["text"])["pos"]
    #neu = analyzer.polarity_scores(target_string)["neu"]
    neg = analyzer.polarity_scores(row["text"])["neg"]
    df.set_value(index, "pos", pos)
    df.set_value(index, "neg", neg)
    df.set_value(index, "compound", compound)


# Print Analysis
df.head(50)
```

```python
df.to_csv("Vader1.csv")
```
